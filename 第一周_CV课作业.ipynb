5.1  Please combine **image crop, color shift, rotation and perspective transform** together to complete a data augmentation script.
   Your code need to be completed in Python/C++ in .py or .cpp file with comments and readme file to indicate how to use.

```python
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt

def my_show(img, size=(3,3)):
    plt.figsize = size
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
def image_crop(img,x1=150, x2=100, y1=100, y2=200): # you code here
    img_ori = cv2.imread(img,1)
    my_show(img_ori[x1:y1][x2:y2])
    pass

def color_shift(img, color = 'Reds'): # you code here
    img_ori = cv2.imread(img,0)
    plt.imshow(img_ori, cmap = color)
    plt.show()    
    pass

def rotation(img, center = (250,250), angel = 30, scale = 1): # you code here
    img_ori = cv2.imread(img,1)
    img = img_ori
    M = cv2.getRotationMatrix2D(center, 30, 1) # center, angle, scale
    img_rotate = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
    cv2.imshow('rotated lenna', img_rotate)
    key = cv2.waitKey(0)
    if key == 27:
        cv2.destroyAllWindows()
    print(M)
    pass

def perspective_transform(img,s1 = [0,0], s2 = [0,500] , s3 =  [500,0], s4 = [500,500], d1 = [5,19], d2 = [19,460], d3 = [460,9], d4 = [410,420]): # you code here
    img_ori = cv2.imread(img, 1)
    pts1 = np.float32([s1,s2,s3,s4])  # 源点创建
    pts2 = np.float32([d1,d2,d3,d4])  # 目标点创建

    M = cv2.getPerspectiveTransform(pts1,pts2)  # 计算得到单应性矩阵
    img_warp = cv2.warpPerspective(img_ori,M,(500,500))  # 通过得到的矩阵对图片进行变换
    my_show(img_warp)
    pass
```

```python
image_crop('lenna.jpg',0,0,100,250)
```

```python
color_shift('lenna.jpg', 'Greens')
```

```python
rotation('lenna.jpg',(250,250),70,0.8)
```

```python
perspective_transform('lenna.jpg')
```
